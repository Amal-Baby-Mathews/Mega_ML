
# Linear regression test

```python
data=pd.read_csv(r"/content/drive/MyDrive/Data_ML/student/student-mat.csv",sep=";")
```

```python
data = data[["G1", "G2","G3", "studytime", "failures", "absences"]]

predict="G3"
```

```python
X = np.array(data.drop([predict], 1))
y= np.array(data[predict])

x_train, x_test, y_train, y_test= sklearn.model_selection.train_test_split(X,y,test_size = 0.1)
```

quick note: linear regression form lines with correlation to data and then use the line for prediction in a dimensional space


```python
linear  = linear_model.LinearRegression()
```

```python
#train the model linear on training data
linear.fit(x_train, y_train)
acc = linear.score(x_test, y_test)
print(acc)
```

```python
#visualize the prediction of the linear model also comment the necessary modules imported
y_pred=linear.predict(x_test)
# display the size of x_test
print(x_test.shape)
print(y_test.shape)
# reshape x_text and then plot it against y_test

plt.scatter(y_test, y_pred, color='blue', label='Actual vs. Predicted')

# Plot the predicted regression line (y = x)
plt.plot(y_pred, y_pred, color='red', linewidth=2, label='Ideal Regression Line')

plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred)')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```
